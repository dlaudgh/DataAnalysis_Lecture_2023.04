```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
titanic = sns.load_dataset("titanic")
```

```python
dt = titanic.pivot_table('survived', 'embark_town', 'sex').round(2)
dt
```

```python
sns.heatmap(dt, cmap='YlOrBr', annot=True, fmt='.2f')
plt.title('타이타닉호의 embark_town/남녀 Heatmap');
```

```python
fig = sns.displot(x='age', kde=True, rug=True, data=titanic)
fig.fig.set_size_inches(5,3)
plt.title('타이타닉호 나이 Dis Plot')
plt.show()
```

```python
filename = '../02.Pandas/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
columns = '구분,총계'.split(',')
cctv = pd.read_csv(filename, encoding='euc=kr', thousands=',')
cctv = cctv[columns]
cctv.drop([0], inplace=True)
cctv.set_index('구분', inplace=True)
cctv.head()
```

```python
sns.barplot(y=cctv.index, x='총계', data=cctv)
plt.title('자치구별 CCTV 총계');
```

```python
si = cctv.총계.sort_index(ascending=False)
si.head()
```

```python
sns.barplot(y=si.index, x=si.values, data=si)
plt.title('자치구별(이름순) CCTV 총계');
```

```python
sv = cctv.총계.sort_values()
sv.head()
```

```python
sns.barplot(y=sv.index, x=sv.values, data=sv)
plt.title('자치구별(총계순) CCTV 총계');
```
