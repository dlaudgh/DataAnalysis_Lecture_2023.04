```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
titanic = sns.load_dataset("titanic")
```

```python
dt = titanic.pivot_table('survived', 'embark_town', 'sex') * 100
dt
```

```python
sns.heatmap(dt, cmap='YlOrBr', annot=True, fmt='.2f')
plt.title('타이타닉호의 embark_town/남녀 생존율 Heatmap');
```

```python
fig = sns.displot(x='age', kde=True, rug=True, data=titanic)
fig.fig.set_size_inches(5,3)
plt.title('타이타닉호 나이 Dis Plot');
```

```python
# 결측치를 평균으로 변경
titanic.age.fillna(titanic.age.mean(), inplace=True)
fig = sns.displot(x='age', kde=True, rug=True, data=titanic)
fig.fig.set_size_inches(5,3)
```

```python
filename = '../02.Pandas/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
columns = '구분,총계'.split(',')
cctv = pd.read_csv(filename, encoding='euc=kr', thousands=',')
cctv = cctv[columns]
cctv.drop([0], inplace=True)
cctv.rename(columns={'구분':'자치구'}, inplace=True)
cctv.head()
```

```python
sns.barplot(x='총계', y='자치구', data=cctv)
plt.title('자치구별 CCTV 총계');
```

```python
si = cctv.sort_values(by='자치구')
sns.barplot(x='총계', y='자치구', data=si)
plt.title('자치구별(이름순) CCTV 총계');
```

```python
sv = cctv.sort_values(by='총계', ascending=False)
sns.barplot(x='총계', y='자치구', data=sv)
plt.title('자치구별 CCTV 총계(총계순)');
```
