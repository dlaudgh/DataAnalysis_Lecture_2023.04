```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg =sns.load_dataset('mpg')
mpg.head()
```

```python
sns.barplot(x='cylinders', y='mpg', data=mpg);
```

```python
sns.scatterplot(x='displacement', y='horsepower', data=mpg);
```

```python
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
mpg.head()
```

```python
df1 = mpg.manufacturer.value_counts().to_frame()
df1.head(10)
```

```python
df2 = mpg.pivot_table('mpg', 'manufacturer', aggfunc='mean').sort_values('mpg', ascending=False).head(10)
df2.head(10)
```

```python
sns.barplot(y=df1.index[:10], x='count', data=df1.head(10));
```

```python
sns.barplot(y=df2.index[:10], x=df2.mpg.head(10));
```
