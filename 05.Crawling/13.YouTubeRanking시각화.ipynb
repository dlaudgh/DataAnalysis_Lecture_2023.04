### YouTube Ranking 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_YouTube랭킹.csv')
df.head(3)
```

- 구독자수 Top 20 - barh 차트로

```python
df1 = df.sort_values(by='구독자수', ascending=False).head(20)
plt.barh(df1.채널명, df1.구독자수)
plt.title('구독자수 Top20')
plt.show()
```

- 조회수 Top 20 - barh 차트로

```python
df2 = df.sort_values(by='조회수', ascending=False).head(20)
plt.barh(df2.채널명, df2.조회수)
plt.title('조회수 Top20')
plt.show()
```

- 비디오수 Top 20 - barh 차트로

```python
df3 = df.sort_values(by='비디오', ascending=False).head(20)
plt.barh(df3.채널명, df3.비디오)
plt.title('비디오수 Top20')
plt.show()
```
